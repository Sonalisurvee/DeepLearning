```python
def combine_date(df,tab_name):
    list_tab = []
    for i in range(df.shape[0]):
        list_tab.append(df.loc[i,'Date'] + 'T' + df.loc[i,tab_name][0:2])
    return np.array(list_tab, dtype = 'datetime64')

df['Datetime'] = combine_date(df,'Time')
df.head
```

```python
def combine_date(df, tab_name):
    list_tab = []
    for i in range(df.shape[0]):
        datetime_str = df.loc[i, 'Date'] + ' ' + df.loc[i, tab_name].replace('.', ':')
        list_tab.append(datetime_str)
    return pd.to_datetime(list_tab, format='%d/%m/%Y %H:%M:%S', errors='coerce')

df['Datetime'] = combine_date(df, 'Time')
```
